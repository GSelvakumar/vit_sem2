#### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Load Dataset

The dataset is collected from Centers of Disease control and prevention [data.cdc.gov](https://www.cdc.gov/) under the category of [Vaccination Coverage among Pregnancy Women](https://data.cdc.gov/Pregnancy-Vaccination/Vaccination-Coverage-among-Pregnant-Women/h7pm-wmjc/data)

```python
# Read the data
data = pd.read_csv('Vaccination_Coverage_among_Pregnant_Women.csv')
```

```python
data['Survey Year/Influenza Season'].unique()
```

```python
# Make a new copy
new_data = data.copy()
```

```python
# Display the presence of missing values
new_data.isnull().sum()
```

```python
new_data.describe()
```

```python
type(new_data['Estimate (%)'][0])
```

```python
new_data['Estimate (%)'].unique()
```

```python
# There is some unwanted values so replacing it with NaN for easy cleaning of data
new_data['Estimate (%)'] = new_data['Estimate (%)'].replace('NR*', 'NaN')
```

```python
new_data['Estimate (%)'].unique()
```

```python
new_data['Estimate (%)'] = new_data['Estimate (%)'].dropna()
```

```python
# The column is in str so typecasting it to float for numerical computations
new_data['Estimate (%)'] = new_data['Estimate (%)'].astype(float)
```

```python
new_data['Estimate (%)'][0]
```

```python
new_data.info()
```

```python
# Dropping all the null values
new_data = new_data.dropna()
```

```python
new_data.info()
```

```python
new_data.describe()
```

```python
new_data.head()
```

```python
# plots the correlations between two variables
ax = sns.scatterplot(data=new_data,
                     x='Estimate (%)',
                     y='Sample Size',
                     )
ax.set_title("Sample Vs Estimate");
```

```python
# plots the correlations between two variables with hue showing the difference between grouping 
ax = sns.scatterplot(data=new_data,
                     x='Estimate (%)',
                     y='Sample Size',
                     hue='Vaccine'
                     )
ax.set_title("Sample Vs Estimate on Vaccines");
```

```python
# plots the correlations between two variables with hue showing the difference between grouping by styling them differently
ax = sns.scatterplot(data=new_data,
                     x='Estimate (%)',
                     y='Sample Size',
                     hue='Vaccine',
                     style='Vaccine',
                     )
ax.set_title("Sample Vs Estimate on Vaccines");
```

```python
# plots the correlations between two variables with hue showing the difference between grouping by styling them differently
ax = sns.scatterplot(data=new_data,
                     x='Estimate (%)',
                     y='Sample Size',
                     hue='Vaccine',
                     style='Vaccine',
                     size='Geography Type'
                     )
ax.set_title("Sample Vs Estimate on Vaccines");
```

```python
# This function provides access to several different axes-level functions that show the relationship between two variables with semantic mappings of subsets
ax = sns.relplot(data=new_data,
                     x='Estimate (%)',
                     y='Sample Size',
                     );
```

```python
ax = sns.relplot(data=new_data,
                     x='Estimate (%)',
                     y='Sample Size',
                     hue='Vaccine',
                     style='Vaccine',
                     size='Survey Year/Influenza Season',
                     sizes=(20, 200),
                     col='Geography Type'
                     );
```

```python
#count plot can be thought of as a histogram across a categorical, instead of quantitative
sns.countplot(x=new_data['Vaccine'])
```

```python
sns.countplot(x=new_data['Vaccine'].sort_values()) #bar in ascending order
```

```python
ax=sns.countplot(x=new_data['Vaccine'].sort_values(ascending=False)) #bar in descending order
ax.set_title("frequency of Vaccines");
```

```python
ax=sns.countplot(y=new_data['Vaccine'].sort_values(ascending=False)) #horizontal bar
ax.set_title("frequency of Vaccines");
```

```python
#bar plot show point estimates and errors as rectangular bars.
sns.barplot(data=new_data,x='Vaccine',y='Sample Size');
```

```python
sns.barplot(data=new_data, y='Vaccine', x='Sample Size'); #horizontal bars
```

```python
sns.barplot(data=new_data, x='Sample Size',y='Vaccine',estimator='max');
```

```python
#histogram that represents the distribution of one or more variables by counting the number of observations that fall within discrete bins.


from matplotlib import pyplot as plt
ax = plt.figure(figsize=(7, 5))
sns.histplot(data = new_data['Survey Year/Influenza Season']);
```

```python
# KDE represents the data using a continuous probability density curve in one or more dimensions.
ax = plt.figure(figsize=(7, 5))
sns.kdeplot(data = new_data['Survey Year/Influenza Season']);
```

```python
# Box Plot shows the distribution of quantitative data with respect to categories
sns.boxplot(data=new_data, x='Survey Year/Influenza Season')
```

```python
# Box Plot shows the distribution of quantitative data with respect to categories
sns.boxplot(data=new_data, 
            x='Survey Year/Influenza Season',
            y ='Geography Type')
```

```python
# Box Plot shows the distribution of quantitative data with respect to categories
sns.boxplot(data=new_data, 
            x='Survey Year/Influenza Season',
            y ='Geography Type',
            hue='Vaccine')
```
