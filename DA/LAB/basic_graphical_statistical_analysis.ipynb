### Plot - Age vs. Price

```python
ax = sns.scatterplot(data=data,
                     x='',
                     y='')
ax.set_title("Sample Size vs Season")
```

```python
ax = sns.scatterplot(data=data,
                     x='Age',
                     y='Price',
                     hue='FuelType',
                     style='FuelType')
ax.set_title("Cars Age vs Price")
```

##### Size is included to alter the visualization 

```python
ax = sns.scatterplot(data=data,
                     x='Age',
                     y='Price',
                     hue='FuelType',
                     style='FuelType',
                     size='KM',
                    )
ax.set_title("Cars Age vs Price")
```

```python
# Custom size is given
ax = sns.scatterplot(data=data,
                     x='Age',
                     y='Price',
                     hue='FuelType',
                     style='FuelType',
                     size='KM',
                     sizes=(20, 250)
                    )
ax.set_title("Cars Age vs Price")
```
